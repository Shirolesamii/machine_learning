```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# import dataset from sklearn
from sklearn.datasets import load_breast_cancer
d = load_breast_cancer()
print(d.keys())
```

```python
# let's build data first
inp = pd.DataFrame(d['data'],columns=d['feature_names'])
inp['target'] = d['target']  #we are adding output
inp.head()
```

```python
# Assignment: Build a Logistic Regression Model, Check the performance as well and list out the your findings
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x=d.data
y=d.target
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
print(y.shape,y_test.shape,y_train.shape)
```

```python
#Model Building
from sklearn.linear_model import LogisticRegression
```

```python
LR=LogisticRegression()
```

```python
LR.fit(X_train,y_train)
```

```python
#testing model
LR.score(X_train,y_train)* 100
```

```python
LR.score(X_test,y_test)*100
```

```python
y_pred=LR.predict(X_test)
y_pred
```

```python
# performance
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(confusion_matrix(y_test,y_pred),fmt='d',xticklabels=["Belign","Malignant"],yticklabels=["Belign","Malignant"],cmap="Reds",annot=True)
plt.xlabel('Belign')
plt.ylabel('Malignant')
plt.show()
```
